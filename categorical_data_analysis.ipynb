```python
import pandas as pd
data={
'Order_ID' : ['01', '02', '03', '04', '05', '06', '07', '08', '09', '010'],

'Category' : ['Electronics', 'clothing', 'electronics', 'grocery','clothing', 'electronics','grocery','clothing', 'electronics', 'grocery'],

"Payment_Method": ["UPI","Card","UPI","Cash","Card","UPI","Cash","Card","UPI","Cash"],

"Delivery_Type": ["Express","Standard","Express","Standard","Express","Standard","Express","Standard","Express","Standard"],

"Customer_Type": ["New","Returning","Returning","New","Returning","New","Returning","New","Returning","New"]
}
df= pd.DataFrame(data)
print(df)
```

```python
#1. frequency count
#count how many orders in each category

print(df['Category'].value_counts())
print()
print(df['Order_ID'].value_counts())
print()
print(df['Payment_Method'].value_counts())
print()
print(df['Delivery_Type'].value_counts())
print()
print(df['Customer_Type'].value_counts())
```

```python
#2. unique values
#Display all unique categories.

print(df['Category'].unique())
```

```python
#3. number of unique values
print(df['Category'].nunique())
```

```python
#4. frequency of payment methods
print(df['Payment_Method'].value_counts())
```

```python
#5. cross tabulation
#category vs payment method
print(pd.crosstab(df['Category'], df['Payment_Method']))
```

```python
#6. cross tabulation (delivery vs customer type)
print(pd.crosstab(df['Delivery_Type'], df['Customer_Type']))
```

```python
#7. percentage distribution
print(df['Category'].value_counts(normalize=True)*100)
```

```python
#8. Filtering data
electronics_orders= df[df['Category']=="Electronics"]
print(electronics_orders)
```

```python
#9. Grouping

print(df.groupby('Category')['Payment_Method'].value_counts())
```

```python
#10. sorting data
sorted_df = df.sort_values(by='Category')
print(sorted_df)
```

```python
#second dataset
import pandas as pd

df=pd.read_csv("/content/Expt11.csv")

print(df.head())
```

```python
#1. count students in each grade
#find how many students obtained each grade

print("Grade Frequency")
print(df['Grade'].value_counts())
```

```python
#2. percentage distribution of grades
print(df['Grade'].value_counts(normalize=True)*100)
```

```python
#3. number of students in each department
print(df['Department'].value_counts())
```

```python
#4. gender distribution
print(df['Gender'].value_counts())
```

```python
#5. gender vs grade analysis
print(pd.crosstab(df['Gender'], df['Grade']))
```

```python
#6. department vs gender
print(pd.crosstab(df['Department'], df['Gender']))
```

```python
#7. department vs grade percentage
print(pd.crosstab(df['Department'], df['Grade'], normalize=True)*100)
```

```python
#8. grouping data (department wise)
print(df.groupby('Department')['Grade'].value_counts())
```
